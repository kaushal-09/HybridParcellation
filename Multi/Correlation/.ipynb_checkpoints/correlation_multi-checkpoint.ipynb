# Calculating the correlation

```python
correlation_labels = None
```

```python
def fetch_features() -> tuple[np.ndarray, list[str]]:
    try:
        # Check if feature file exists
        feat_file = feature_filename
        X_features = np.load(feat_file)['a']
        labels = np.load(feat_file,allow_pickle = True)['labels']
        print("Feature file found.")
        return X_features, labels
    except FileNotFoundError:
        # if not then return None
        return None, None
```

```python
def get_raw_atlas(atlas: str) -> tuple[str, any, list[str]]:
    dictionary = {
        "basc": ("ABIDE_BASC064.npz",
                 datasets.fetch_atlas_basc_multiscale_2015().scale064,
                 None),
        "harvard_oxford": ("ABIDE_HARVARD_OXFORD.npz",
                           datasets.fetch_atlas_harvard_oxford("cort-maxprob-thr25-2mm").maps,
                           datasets.fetch_atlas_harvard_oxford("cort-maxprob-thr25-2mm").labels),
        "aal": ("ABIDE_AAL.npz",
                datasets.fetch_atlas_aal().maps,
                datasets.fetch_atlas_aal().labels)
    }

    if val := dictionary.get(atlas):
        feature_filename, atlas_filename, labels = val
        feature_filename = os.path.join(output_dir, feature_filename)
        return feature_filename, atlas_filename, labels

    return None, None, None
```

```python
def make_feature_file():
    global correlation_labels
    X_features = []
    print("No feature file found. Extracting features...")

    fmri_filenames = abide['func_preproc']

    # load atlas
    
    multiscale = datasets.fetch_atlas_basc_multiscale_2015()
    atlas_filename = multiscale.scale064
    
    #atlas = datasets.fetch_atlas_harvard_oxford("cort-maxprob-thr25-2mm")
    #atlas_filename = atlas.maps
    #labels = atlas.labels
    
    #feature_fname, atlas_filename, labels = get_raw_atlas("aal")

    masker = NiftiLabelsMasker(labels_img=atlas_filename,
                               standardize=True,
                               memory='nilearn_cache',
                               verbose=0)

    # initialize correlation measure
    correlation_measure = ConnectivityMeasure(kind='correlation', vectorize=True,
                                              discard_diagonal=True)
    #'correlation', 'covariance', 'partial correlation', 'tangent', 'precision'

    for i, sub in enumerate(fmri_filenames):
        # extract the timeseries from the ROIs in the atlas
        time_series = masker.fit_transform(sub)
        # create a region x region correlation matrix
        correlation_matrix = correlation_measure \
            .fit_transform([time_series])[0]
        X_features.append(correlation_matrix)
        print(f'finished extracting {i + 1} of {len(fmri_filenames)}')
    
    # Save features
    np.savez_compressed(
        feature_filename,
        a=X_features,
        labels=labels)
```

```python
X_features, labels = fetch_features()

if X_features is None:
    make_feature_file()
    X_features, labels = fetch_features()
```

# Models

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier, ExtraTreesClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB, BernoulliNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
```

```python
#Decision Tree Classifier
dt_model = DecisionTreeClassifier(min_samples_leaf=10, max_depth=100, criterion='entropy', splitter='best', max_leaf_nodes=1000, min_impurity_decrease=1e-5)

#Bagging Classifier
bg_model = BaggingClassifier(estimator=LogisticRegression(max_iter=10000, C=1000), n_estimators=10000, max_samples=0.95, max_features=0.95, bootstrap=True, bootstrap_features=False, warm_start=True)

#Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=10000, max_depth=100, max_features='sqrt', criterion='entropy', bootstrap=True, oob_score=True, warm_start=True, min_samples_split=5)

#Support Vector Machine Classifier
sv_model = SVC(C=100, kernel='rbf', gamma='scale', probability=True, shrinking=True, cache_size=5000)

#Logistic Regression Classifier
lr_model = LogisticRegression(max_iter=10000, C=10000, solver='lbfgs', penalty='l2', dual=False, fit_intercept=True, intercept_scaling=1, multi_class='ovr')

#Gradient Boosting Classifier
gbm_model = GradientBoostingClassifier(n_estimators=500, max_depth=50, learning_rate=0.05, subsample=0.95, min_samples_split=5, min_samples_leaf=5, max_features='sqrt')

#AdaBoost Classifier
adaboost_model = AdaBoostClassifier(n_estimators=1000, learning_rate=0.1, algorithm='SAMME.R')

#K-Nearest Neighbors Classifier
knn_model = KNeighborsClassifier(n_neighbors=20, weights='distance', p=2, algorithm='ball_tree', leaf_size=30)

#Gaussian Naive Bayes Classifier
nb_model = GaussianNB(var_smoothing=1e-15)

#Bernoulli Naive Bayes Classifier
bnb_model = BernoulliNB(alpha=0.05, binarize=None)

#Extra Trees Classifier
et_model = ExtraTreesClassifier(n_estimators=10000, max_depth=100, max_features='sqrt', criterion='entropy', bootstrap=True, oob_score=True, warm_start=True, min_samples_split=5)

#XGBoost Classifier
xgb_model = XGBClassifier(n_estimators=500, max_depth=50, learning_rate=0.05, subsample=0.95, objective='binary:logistic', tree_method='hist', grow_policy='lossguide', max_leaves=1000)

#Linear Discriminant Analysis
lda_model = LinearDiscriminantAnalysis(solver='lsqr', shrinkage='auto')

#Neural Network Classifier
nn_model = MLPClassifier(hidden_layer_sizes=(1000, 500, 200, 100), max_iter=10000, activation='relu', solver='adam', alpha=0.0001, batch_size=500, learning_rate='adaptive', learning_rate_init=0.001, momentum=0.9)

dt_model.fit(xtrain, ytrain)
bg_model.fit(xtrain, ytrain)
rf_model.fit(xtrain, ytrain)
sv_model.fit(xtrain, ytrain)
lr_model.fit(xtrain, ytrain)
gbm_model.fit(xtrain, ytrain)
adaboost_model.fit(xtrain, ytrain)
knn_model.fit(xtrain, ytrain)
nb_model.fit(xtrain, ytrain) 
#xgb_model.fit(xtrain, ytrain)
lda_model.fit(xtrain, ytrain)
bnb_model.fit(xtrain, ytrain)
et_model.fit(xtrain, ytrain)
nn_model.fit(xtrain, ytrain)

dt_preditions = dt_model.predict(xtest)
bg_preditions = bg_model.predict(xtest)
rf_preditions = rf_model.predict(xtest)
sv_preditions = sv_model.predict(xtest)
lr_preditions = lr_model.predict(xtest)
gbm_predictions = gbm_model.predict(xtest)
ada_predictions = adaboost_model.predict(xtest)
knn_predictions = knn_model.predict(xtest)
nb_predictions = nb_model.predict(xtest)
#xgb_predictions = xgb_model.predict(xtest)
lda_predictions = lda_model.predict(xtest)
bnb_predictions = bnb_model.predict(xtest)
et_predictions = et_model.predict(xtest)
nn_predictions = nn_model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

models = [("Decision Tree", dt_model),
          ("Bagging Classifier", bg_model),
          ("Logistic Regression", lr_model),
          ("Random Forest", rf_model),
          ("Support Vector Machine", sv_model),
          ("Gradient Boosting Machines", gbm_model),
          ("AdaBoost", adaboost_model),
          ("K-Nearest Neighbors", knn_model),
          ("Naive Bayes", nb_model),
          ("Neural Networks", nn_model),
         ("LinearDiscriminantAnalysis", lda_model),
         ("BernoulliNB", bnb_model),
         ("ExtraTreesClassifier", et_model)]

accs = []
for name, model in models:
    y_pred = model.predict(X_features)
    tn, fp, fn, tp = confusion_matrix(y_target, y_pred).ravel()
    accuracy = accuracy_score(y_target, y_pred)
    sensitivity = tp / (tp + fn)
    specificity = tn / (tn + fp)
    accs.append((name, accuracy, sensitivity, specificity))

accuracy_df = pd.DataFrame(accs, columns=['Model', 'Accuracy', 'Sensitivity', 'Specificity'])
atlas_used, measure_used = os.getcwd().split(os.sep)[-2:]
accuracy_df.to_csv(f"./XFeatures-{atlas_used}-{measure_used}.csv")
accuracy_df = accuracy_df.sort_values("Accuracy", ascending=False).reset_index(drop=True)
accuracy_df
```
