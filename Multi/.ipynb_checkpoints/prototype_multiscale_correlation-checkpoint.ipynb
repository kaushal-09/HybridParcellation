# Calculating the correlation

```python
correlation_labels = None
```

```python
def fetch_features() -> tuple[np.ndarray, list[str]]:
    try:
        # Check if feature file exists
        feat_file = feature_filename
        X_features = np.load(feat_file)['a']
        labels = np.load(feat_file, allow_pickle=True)['labels']
        print("Feature file found.")
        return X_features, labels
    except FileNotFoundError:
        # if not then return None
        return None, None
```

```python
def get_raw_atlas(atlas: str) -> tuple[str, any, list[str]]:
    dictionary = {
        "basc": ("ABIDE_BASC064.npz",
                 datasets.fetch_atlas_basc_multiscale_2015().scale064,
                 None),
        "harvard_oxford": ("ABIDE_HARVARD_OXFORD.npz",
                           datasets.fetch_atlas_harvard_oxford("cort-maxprob-thr25-2mm").maps,
                           datasets.fetch_atlas_harvard_oxford("cort-maxprob-thr25-2mm").labels),
        "aal": ("ABIDE_AAL.npz",
                datasets.fetch_atlas_aal().maps,
                datasets.fetch_atlas_aal().labels)
    }

    if val := dictionary.get(atlas):
        feature_filename, atlas_filename, labels = val
        feature_filename = os.path.join(output_dir, feature_filename)
        return feature_filename, atlas_filename, labels

    return None, None, None
```

```python

def make_feature_file():
    global correlation_labels
    X_features = []
    print("No feature file found. Extracting features...")

    fmri_filenames = abide['func_preproc']

    # load atlas
    # multiscale = datasets.fetch_atlas_basc_multiscale_2015()
    # atlas_filename = multiscale.scale064
    # atlas = datasets.fetch_atlas_harvard_oxford("cort-maxprob-thr25-2mm")
    # atlas_filename = atlas.maps
    feature_fname, atlas_filename, labels = get_raw_atlas("basc")

    masker = NiftiLabelsMasker(labels_img=atlas_filename,
                               standardize=True,
                               memory='nilearn_cache',
                               verbose=0)

    # initialize correlation measure
    correlation_measure = ConnectivityMeasure(kind='correlation', vectorize=True,
                                              discard_diagonal=True)

    for i, sub in enumerate(fmri_filenames):
        # extract the timeseries from the ROIs in the atlas
        time_series = masker.fit_transform(sub)
        
        # create a region x region correlation matrix
        correlation_matrix = correlation_measure \
            .fit_transform([time_series])[0]
        
        correlation_matrix = fft(correlation_matrix)
        # correlation_matrix = [val.imag for val in correlation_matrix]

        X_features.append(correlation_matrix)
        print(f'finished extracting {i + 1} of {len(fmri_filenames)}')
    
    # Save features
    np.savez_compressed(
        feature_fname,
        a=X_features,
        labels=labels)
```

## Principle Component Analysis

```python
X_features_pca = PCA(0.99).fit_transform(X_features.imag) # keeping 99% of variance

# Transform phenotypic data into dataframe
abide_pheno = pd.DataFrame(abide.phenotypic)

# Get the target vector
y_target = abide_pheno['DX_GROUP']

```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X_features_pca, y_target, test_size=0.15, shuffle=True)
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape 
```

# Models

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
dt_model = DecisionTreeClassifier(min_samples_leaf=4, max_depth=15)
bg_model = BaggingClassifier(estimator=LogisticRegression(), n_estimators=1000)
rf_model = RandomForestClassifier(n_estimators=1000, max_depth=15)
sv_model = SVC(C=3, kernel='rbf')
lr_model = LogisticRegression()
```

```python
dt_model.fit(xtrain, ytrain)
bg_model.fit(xtrain, ytrain)
rf_model.fit(xtrain, ytrain)
sv_model.fit(xtrain, ytrain)
lr_model.fit(xtrain, ytrain)
```

```python
dt_preditions = dt_model.predict(xtest)
bg_preditions = bg_model.predict(xtest)
rf_preditions = rf_model.predict(xtest)
sv_preditions = sv_model.predict(xtest)
lr_preditions = lr_model.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

## Result

```python
models = [("Decision Tree", dt_model),
          ("Bagging Classifier", bg_model),
          ("Logistic Regression", lr_model),
          ("Random Forest", rf_model),
          ("Support Vector Machine", sv_model)]
accs = [(name, accuracy_score(ytest, model.predict(xtest))) for name, model in models]

accuracy_df = pd.DataFrame(accs, columns=['Model', 'Accuracy'])
accuracy_df.sort_values("Accuracy", ascending=False).reset_index(drop=True)
```
