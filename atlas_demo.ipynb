```python
from nilearn import datasets
```

```python
atlas_ho = datasets.fetch_atlas_harvard_oxford("cort-maxprob-thr25-2mm")
atlas_aal = datasets.fetch_atlas_aal()
atlas_basc_multiscale = datasets.fetch_atlas_basc_multiscale_2015()
```

```python
from nilearn import plotting
from matplotlib import pyplot as plot
```

# Harvard Oxford Atlas Plotting

```python
plotting.plot_roi(atlas_ho.filename, title="Harvard Oxford atlas", figure=plot.figure(1, (15, 7)))
```

# AAL Atlas

```python
plotting.plot_roi(atlas_aal.maps, title="AAL atlas", figure=plot.figure(1, (15, 7))) 
```

# BASC Multiscale

```python
plotting.plot_roi(atlas_basc_multiscale.scale064, title="BASC Multiscale atlas", figure=plot.figure(1, (15, 7))) 
```
